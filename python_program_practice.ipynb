```python
#List is a data type, which is ordered, mutable ,contains duplicates,conatins other data types as its element
name = "Hello"
print(name)

fruits = ["apple", "mangoe", "guava", "guava"]
print(fruits)
#retrieveing values from list using indexing
print("First fruit is:",fruits[0])
print("last fruit is:", fruits[-1])
print("adding items to fruit")
fruits.append("grapes")
fruits
fruits.remove("apple")
fruits

```

```python
#Tuple is a data type , which is ordered ,immutable, allow duplicate values and contains other datatypes as its elelment
my_tuple=("Hi",1,True,1.7)
my_tuple

#retrieveing values from tuple using indexing
print(my_tuple[-1])
print(my_tuple[1])

#unpacking in tuple : getting individual elements of the tuple, can read the tuple elents
numbers=(10,20,300,90)
a,b,c,d=numbers
print(a)
print(b)
print(c)
print(d)

#iterating
tup=(1,2,3)
for x in tup:
  print(x)

#combining two tuples : concatination
tup1=(1,2,3,4,5)
tup2=(100,300,45,90)
print(tup1 + tup2)

#deleting the tuple
tup=(3,4,5)
del(tup)
#tup




```
